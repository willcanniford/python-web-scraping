```python
# Imports
import requests
import pandas as pd
from bs4 import BeautifulSoup
import re 
import codecs
import json
import matplotlib.pyplot as plt
```

```python
pattern = r".*var\sstat_data\s+=\s+\$\.parseJSON\(\"(.*?)\"\)"
url = 'https://ighthamcc.play-cricket.com/website/player_stats_widget/batting_stats/307896'
html = requests.get(url)
soup = BeautifulSoup(html.text, "html.parser")
scripts = soup.find_all("script")
# Isolate the script with the relevant data
data_script = [i.string for i in scripts if i.string is not None and 'stat_data' in i.string][0]
pattern = re.compile(pattern)
bytes_data = codecs.escape_decode(re.search(pattern, data_script.string).group(1))
json_data = json.loads(bytes_data[0].decode("utf-8"))
json_data.keys()
```

```python
df = pd.Series(json_data['Division']).apply(pd.Series)
df.set_index('season')
```
