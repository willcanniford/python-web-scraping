```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
import re 
import codecs
import json
pattern = r"shotsData\s+=\s+JSON.parse\(\'(.*?)\'\)"
html = requests.get("https://understat.com/match/12006/")
soup = BeautifulSoup(html.text, "html.parser")
scripts = soup.find_all("script")
pattern = re.compile(pattern)
bytes_data = codecs.escape_decode(re.search(pattern, scripts[1].string).group(1))
json_data = json.loads(bytes_data[0].decode("utf-8"))
json_data.keys()
pd.DataFrame.from_dict(json_data).h.apply(pd.Series).append(pd.DataFrame.from_dict(json_data).a.apply(pd.Series))
```
